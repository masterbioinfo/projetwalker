```python
###################################################################################################################

                                            ### Dependencies ###

###################################################################################################################

from ipywidgets import *
import ipywidgets as widgets
from __future__ import print_function
from IPython.display import display, clear_output
from IPython.html import widgets
import os


#####################################3D structure dependencies
import py3Dmol

###################################################################################################################

                                            ### Classes ###

###################################################################################################################

class Titration(object):
    """Titration class test."""
    def __init__(self, checkPlot, plotName, plotValue):
        self.checkPlot = checkPlot
        self.plotName = plotName
        self.plotValue = plotValue

###################################################################################################################

                                            ### Functions ###

###################################################################################################################
    
def parametersDefault(button):
    """ """
    
    #return default.value = True
    
def validateSelection(button):
    """Change parameters selected by the user when he push validate button."""
    global selectResiduList
    global notRetainedList
    global selectResiduCache
    global notRetainedResiduCache
    
    if default.value == False :
        
        if selectResidu.value != "0" and selectResidu.value != "":
            line = selectResidu.value.split("-")
            for residu in line:
                selectResiduList.append(int(residu))
        elif selectResidu.value == "0":
            selectResiduList = []
            
        if notSelectResidu.value != "0" and selectResidu.value != "":
            line = notSelectResidu.value.split("-")
            for residu in line:
                notRetainedList.append(int(residu))
        elif notSelectResidu.value == "0":
            notRetainedList = []
        
        pdbCode.value = pdbCode.value
        proteinRepresentation.value = proteinRepresentation.value
        colorProtein.value = colorProtein.value
            
        selectResidu.value = ""
        residuRepresentation.value = residuRepresentation.value
        colorResidu.value = colorResidu.value
    
        notSelectResidu.value = ""
        notSelectedResiduRepresentation.value = notSelectedResiduRepresentation.value
        colorResiduNotRetained.value = colorResiduNotRetained.value
    
        colorBackGround.value = colorBackGround.value
        
        view3D(selectResiduList,notRetainedList)
        
        consoleDisplay()
    
    elif default.value == True :

        selectResiduList = selectResiduCache
        notRetainedList = notRetainedResiduCache
        
        pdbCode.value = pdbCode.value
        proteinRepresentation.value = "cartoon"
        colorProtein.value = "#FFFFFF"
            
        selectResidu.value = ""
        residuRepresentation.value = "stick"
        colorResidu.value = "#FF000D"
    
        notSelectResidu.value = ""
        notSelectedResiduRepresentation.value = "stick"
        colorResiduNotRetained.value = "#09FF0B"
    
        colorBackGround.value = "#FFFFFF"
        view3D(selectResiduList,notRetainedList)
        
        consoleDisplay()
        default.value = False

def consoleDisplay():
    """Display all modifications done in 3D representation menu."""
    
    if default.value == False :
        console.value = "Residus selected : {0}.\nResidus not selected : {1}.".format(str(selectResiduList).strip("[").strip("]"), str(notRetainedList).strip("[").strip("]"))
    elif default.value == True :
        console.value = "All parameters return at defaults values."
            
def notRetained(notRetainedList):
    """Display in the console all residus not retained in the study."""
    
    warningPhrase = "Residus not retained in the study: \n"
    if len(notRetainedList) == 0:
        residuNotReatined.value = "All residus retained in the study !"
    elif len(notRetainedList) > 0 :
        for residu in notRetainedList :
            residu = str(residu)
            if warningPhrase == "Residus not retained in the study: \n":
                warningPhrase = "{0}{1}".format(warningPhrase, residu)
            elif warningPhrase != "Residus not retained in the study: \n":
                warningPhrase = "{0}, {1}".format(warningPhrase, residu)
    
    return warningPhrase   

###################################################################################################################

                                        ### Python code not graphical ###

###################################################################################################################

#####################################Residu selected

# Warning : change selectResiduCache by residus over the cutoff. 
#Change notRetainedResiduCache by residus not retained in the study. 

selectResiduCache = [50, 51]
notRetainedResiduCache = [123, 142, 202]

selectResiduList = selectResiduCache
notRetainedList = notRetainedResiduCache
warningPhrase = notRetained(notRetainedList)

#####################################Plot list
plotList = list()
i = 0
while i <= 10 : 
    plot = Titration(False, "Step {0}".format(i), i)
    plotList.append(plot)
    i += 1
    
plot = Titration(False, "All steps", "all")
plotList.append(plot) 

stepList=list()
for plot in plotList:
    stepList.append(plot.plotName)
    
###################################################################################################################

                                            ### Widgets ###

###################################################################################################################

#####################################Principal menu

default = widgets.Checkbox(value=False, description='Parameters by default',disabled=False, 
        tooltip='Return parameters values by default')
validate = widgets.Button(value=False, description='Validate', disabled=False, button_style='', 
        tooltip='Validate plot and cutoff choice', icon='check')

default.observe(parametersDefault)
validate.on_click(validateSelection)

#########################################Menu 3D structure

pdbCode = widgets.Text(placeholder='Enter PDB code, exemple: 4wk4',description='PDB code', value="0",
    disabled=False)
proteinRepresentation = widgets.Dropdown(options=['Initial', 'cartoon', 'cross', 'line', 'stick', 'sphere'], 
    description='Protein representation', disabled=False,)
colorProtein = widgets.ColorPicker(concise=True, description='Protein color', disabled=False)

selectResidu = widgets.Text(placeholder='Enter residu number',description='Residu number', value="0",
    disabled=False)
residuRepresentation = widgets.Dropdown(options=['Initial', 'cartoon', 'cross', 'stick', 'line', 'sphere'],
    description='Residu representation', disabled=False,)
colorResidu = widgets.ColorPicker(concise=True, description='Residus color', disabled=False)

notSelectResidu = widgets.Text(placeholder='Enter residu not retained number',description='Residu number', value="0",
    disabled=False)
notSelectedResiduRepresentation = widgets.Dropdown(options=['Initial', 'cartoon', 'cross', 'line', 'stick', 'sphere'],
    description='Residu not retained representation', disabled=False,)
colorResiduNotRetained = widgets.ColorPicker(concise=True, description='Residus not retained color', disabled=False)

colorBackGround = widgets.ColorPicker(concise=True, description='Backgorund', 
    disabled=False)


#Variable initialisaton

if pdbCode.value == "0" :
    pdbCode.value = "4kw4"
if colorProtein.value == "black":
    colorProtein.value = '#FFFFFF'
if proteinRepresentation.value == "Initial" :
    proteinRepresentation.value = "cartoon"

if selectResidu.value == "0":
    selectResidu.value = ""
if residuRepresentation.value == "Initial":
    residuRepresentation.value = "stick"
if colorResidu.value == "black":
    colorResidu.value = "#FF000D"

if notSelectResidu.value == "0":
    notSelectResidu.value = ""
if notSelectedResiduRepresentation.value == "Initial":
    notSelectedResiduRepresentation.value = "stick"
if colorResiduNotRetained.value == "black":
    colorResiduNotRetained.value = "#09FF0B"
    
if colorBackGround.value == "black":
    colorBackGround.value = "#FFFFFF"
    
#########################################Console

residuNotRetained = widgets.Textarea(value= notRetained(notRetainedList), placeholder='Residus not retained in the study', description='', disabled=False, 
    display='flex', flex_flow='row', height='1000px')
console = widgets.Textarea(value= '', placeholder='Console', description='', disabled=False, 
    display='flex', flex_flow='row', height='1000px')

consoleDisplay()
consolesMenu = HBox([residuNotRetained, console])



###################################################################################################################

                                        ### Display widgets ###

###################################################################################################################

#########################################Organisation

#Third box level
lefter_box = VBox([pdbCode, selectResidu, notSelectResidu])
center_box = VBox([proteinRepresentation, residuRepresentation, notSelectedResiduRepresentation])
righter_box = VBox([colorProtein, colorResidu,colorResiduNotRetained, colorBackGround])
menuStructure = HBox([lefter_box, center_box, righter_box])

principalMenu = VBox([default, validate])

#Secondary box level
center_box = HBox([menuStructure])
lower_box = HBox([principalMenu, consolesMenu])

#Primary box level
VBox([center_box, lower_box], min_height='100%', min_width='100%')
```

```python
###################################################################################################################

                                        ### 3D Representation ###

###################################################################################################################

def view3D(selectResiduList,notRetainedList):
    """Show 3D representation protein."""
    ########################################3D structure    
    #support for pdb, sdf, mol2, xyz, and cube formats
    #parallelized molecular surface computation
    #sphere, stick, line, cross, cartoon, and surface styles
    #atom property based selection and styling labels
    #clickable interactivity with molecular data
    #geometric shapes including spheres and arrows

    #Protein
    print("Code PDB: {0}".format(pdbCode.value))
    protein = pdbCode.value
    pdb = 'pdb:' + protein
    view = py3Dmol.view(query=pdb)
    #color protein by default 'white'
    view.setStyle({proteinRepresentation.value:{'color':colorProtein.value}})

    #Background
    backgroundColor = "0x{0}".format('ffffff'.strip("#"))

    #Residus selected with histograms selection and residus not selectionned in the study..
    residus = {'residuSelected': selectResiduList, 'notRetained': notRetainedList }
    color = {'residuSelected': 'red','notReatined': 'green' }

    #Draw residu retained in red.
    res = {'resi': residus['residuSelected']}
    view.setStyle(res,{residuRepresentation.value:{'color':colorResidu.value}})
    #Draw residu retained in green.
    res = {'resi': residus['notRetained']}
    view.setStyle(res,{notSelectedResiduRepresentation.value:{'color':colorResiduNotRetained.value}})
    #Select a background.
    view.setBackgroundColor(backgroundColor)
    #Display 3D structure.
    display(view)

view3D(selectResiduList,notRetainedList)
```
